```python
import nltk
from nltk.corpus import stopwords
import string

# Download required NLTK resources (run once if you haven't already)
# nltk.download('stopwords')
# Note: 'punkt' download has been removed to avoid punkt_tab issues

def preprocess_text(paragraph):
    """
    Preprocesses a given paragraph by lowercasing, removing punctuation, 
    and removing stopwords without using NLTK's punkt tokenizer.
    """
    
    # 1. Lowercase the text
    lowercased_text = paragraph.lower()
    
    # 2. Remove punctuation
    # Create a translation table that maps punctuation characters to None
    translator = str.maketrans('', '', string.punctuation)
    no_punctuation_text = lowercased_text.translate(translator)
    
    # 3. Tokenize the text (split into words)
    # Using Python's built-in split() instead of NLTK's word_tokenize to avoid punkt
    words = no_punctuation_text.split()
    
    # 4. Remove stopwords
    # Get the list of English stopwords from NLTK
    stop_words = set(stopwords.words('english'))
    
    # Keep only the words that are not in the stopwords list
    filtered_words = [word for word in words if word not in stop_words]
    
    # 5. Rejoin the words back into a string (optional, but usually desired)
    preprocessed_paragraph = " ".join(filtered_words)
    
    return preprocessed_paragraph

# Example usage
sample_paragraph = "Hello there! This is a sample paragraph. It contains some punctuation, and several stopwords that we want to remove. Let's see how it works out."

print("Original:")
print(sample_paragraph)
print("\nPreprocessed:")
print(preprocess_text(sample_paragraph))
```

```python
#Convert Text Corpus to TF-IDF Matrix
from sklearn.feature_extraction.text import TfidfVectorizer

corpus = [
    "I love machine learning",
    "Machine learning is fun",
    "Python is powerful",
    "I love Python"
]

vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(corpus)

print("Vocabulary:")
print(vectorizer.get_feature_names_out())

print("\nTF-IDF Matrix:")
print(tfidf_matrix.toarray())
```

```python
#Train Naive Bayes on SMS Spam Dataset
import pandas as pd

data = {
    "label": [
        "ham",
        "spam",
        "ham",
        "spam",
        "ham",
        "spam",
        "ham",
        "spam"
    ],
    "message": [
        "How are you?",
        "Congratulations! You won a lottery",
        "Let's meet tomorrow",
        "Claim your free prize now",
        "Call me later",
        "You have won cash reward",
        "See you in class",
        "Free entry in lucky draw"
    ]
}

df = pd.DataFrame(data)

df.to_csv("spam.csv", index=False)

print("spam.csv created successfully!")

#Step 2: Train Naive Bayes

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

# Load dataset
df = pd.read_csv("spam.csv")

X = df["message"]
y = df["label"]

# Convert text into numerical features
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(X)

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.25,
    random_state=42
)

# Train model
model = MultinomialNB()
model.fit(X_train, y_train)

# Predict
predictions = model.predict(X_test)

print("Predictions:")
print(predictions)
```

```python
#Classification Report and Confusion Matrix
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, confusion_matrix

# Load dataset
df = pd.read_csv("spam.csv")

X = df["message"]
y = df["label"]

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.25,
    random_state=42
)

model = MultinomialNB()
model.fit(X_train, y_train)

predictions = model.predict(X_test)

print("Classification Report")
print(classification_report(y_test, predictions))

print("Confusion Matrix")
print(confusion_matrix(y_test, predictions))
```

```python
#1. Movie Review Sentiment Classifier (Positive vs Negative)
#Step 1: Create Dataset (movie_reviews.csv)
import pandas as pd

data = {
    "review": [
        "This movie is amazing",
        "I loved the film",
        "Excellent acting and story",
        "Fantastic experience",
        "Very good movie",
        "Worst movie ever",
        "I hated this film",
        "Terrible acting",
        "Very boring movie",
        "Waste of time"
    ],
    "sentiment": [
        "Positive",
        "Positive",
        "Positive",
        "Positive",
        "Positive",
        "Negative",
        "Negative",
        "Negative",
        "Negative",
        "Negative"
    ]
}

df = pd.DataFrame(data)
df.to_csv("movie_reviews.csv", index=False)

print("movie_reviews.csv created successfully!")

#Step 2: Train Sentiment Classifier
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, confusion_matrix

# Load dataset
df = pd.read_csv("movie_reviews.csv")

X = df["review"]
y = df["sentiment"]

# TF-IDF
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(X)

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Train model
model = MultinomialNB()
model.fit(X_train, y_train)

# Prediction
pred = model.predict(X_test)

print("Predictions:")
print(pred)

print("\nClassification Report")
print(classification_report(y_test, pred))

print("\nConfusion Matrix")
print(confusion_matrix(y_test, pred))

# Test your own review
review = ["This movie is wonderful and fantastic"]
review_vector = vectorizer.transform(review)

prediction = model.predict(review_vector)

print("\nNew Review Prediction:", prediction[0])
```

```python
#2. News Category Classifier (TF-IDF + Logistic Regression)
#Step 1: Create Dataset (news.csv)
import pandas as pd

data = {
    "news": [
        "India won the cricket match",
        "New AI technology launched",
        "Stock market reaches record high",
        "Football World Cup begins",
        "Government announces new policy",
        "Scientists discover new planet",
        "Basketball team wins championship",
        "Election results declared"
    ],
    "category": [
        "Sports",
        "Technology",
        "Business",
        "Sports",
        "Politics",
        "Science",
        "Sports",
        "Politics"
    ]
}

df = pd.DataFrame(data)

df.to_csv("news.csv", index=False)

print("news.csv created successfully!")

#Step 2: Train Logistic Regression Model
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

# Load dataset
df = pd.read_csv("news.csv")

X = df["news"]
y = df["category"]

# TF-IDF Vectorization
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(X)

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.25,
    random_state=42
)

# Logistic Regression
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# Prediction
pred = model.predict(X_test)

print("Predictions:")
print(pred)

print("\nClassification Report")
print(classification_report(y_test, pred))

print("\nConfusion Matrix")
print(confusion_matrix(y_test, pred))

# Test custom news
sample = ["The Indian cricket team won the final match"]

sample_vector = vectorizer.transform(sample)

prediction = model.predict(sample_vector)

print("\nPredicted Category:", prediction[0])
```
