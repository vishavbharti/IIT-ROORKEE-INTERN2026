```python
phonebook = {}
phonebook["A"] = "123-123"
phonebook["B"] = "9876-5432"
phonebook["C"] = "1234-5678"
print("Phonebook after adding:", phonebook)
search_name = "B"
number = phonebook.get(search_name, "Contact not found")
print(f"{search_name}'s number is: {number}")
rmv_number = phonebook.pop("C", "Not found")
print(f"Deleted C (Number was {rmv_number})")
print("Final Phonebook:", phonebook)
```

```python
sentence = "she sells, sea shells, by the sea shore."
words = sentence.split()
wordcount = {}
for word in words:
    if word in wordcount:
        wordcount[word] += 1
    else:
        wordcount[word] = 1
print("Word Frequencies:")
for word, count in wordcount.items():
    print(f"{word}: {count}")
```

```python
text = "hello world"
chars = set(text)
print(f"Original text: '{text}'")
print(f"Unique characters: {chars}")
if " " in chars:
    chars.remove(" ")
print(f"Original text: '{text}'")
print(f"Unique characters(no spaces): {chars}")
```

```python
def palindrome(word):
    my_word = word.strip().lower()
    reversed_word = my_word[::-1]
    
    print(f"Original: {my_word}")
    print(f"Reversed: {reversed_word}")
    
    if my_word == reversed_word:
        print("Result: It is a palindrome!\n")
    else:
        print("Result: Not a palindrome.\n")

palindrome("Racecar")
palindrome("Python")
```

```python
report = {
    "A": {"marks": 85, "grade": "B"},
    "B": {"marks": 92, "grade": "A"},
    "C": {"marks": 65, "grade": "D"}
}
report["D"] = {"marks": 78, "grade": "C"}
print("\nFull Report Card:")
for student, info in report.items():
    print(f"Student: {student}, Marks: {info['marks']}, Grade: {info['grade']}")
n = input('Enter student\'s name: ').upper()
print(f"{n}\'s Marks and Grades are:", report[n]["marks"],"and", report[n]["grade"], "respectively.")
```

```python
def anagram(word1, word2):
    w1 = sorted(word1)
    w2 = sorted(word2)
    if w1 == w2:
        print(f"'{word1}' and '{word2}' are anagrams.")
    else:
        print(f"'{word1}' and '{word2}' are not anagrams.")
```

```python
anagram("listen", "silent")
anagram("hello", "world")
```

```python
def caesar(text, shift, mode="encode"):
    result = ""
    if mode == "decode":
        shift = -shift
    for char in text:
        if char.isalpha():
            start = ord('A') if char.isupper() else ord('a')
            shifted = (ord(char) - start + shift) % 26
            result += chr(shifted + start)
        else:
            result += char           
    return result
secret_message = caesar("Hello World!", 8, "encode")
print("Encoded:", secret_message)

decoded_message = caesar(secret_message, 8, "decode")
print("Decoded:", decoded_message)
```
