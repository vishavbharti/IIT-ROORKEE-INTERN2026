```python
#iris data set
import pandas as pd
from sklearn.datasets import load_iris

# Load the dataset
iris = load_iris()

# Create a DataFrame for easier exploration
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target

# Map numerical targets to actual species names for readability
species_map = {i: name for i, name in enumerate(iris.target_names)}
df['species_name'] = df['species'].map(species_map)
```

```python
print(df.describe())
```

```python
print(df['species_name'].value_counts())
```

```python
#Visualise feature distributions using histograms and pair plots
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set_theme(style="whitegrid")

# Plot Histograms for each feature
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
features = iris.feature_names

for i, col in enumerate(features):
    ax = axes[i//2, i%2]
    sns.histplot(data=df, x=col, hue="species_name", kde=True, ax=ax, element="step")
    ax.set_title(f'Distribution of {col}')

plt.tight_layout()
plt.show()
```

```python
# Generate a pair plot colored by the species
sns.pairplot(df.drop(columns=['species']), hue="species_name", palette="Set2")
plt.show()
```
